```python
face_cascade = cv2.CascadeClassifier(r'C:/Users/Gaya/Desktop/Computer-Vision-with-Python/DATA/haarcascades/haarcascade_frontalface_default.xml')
```

```python
def detect_face(img):
    
    face_img = img.copy()
    
    face_rects = face_cascade.detectMultiScale(face_img)
    
    for (x,y,w,h) in face_rects:
        
        cv2.rectangle(face_img,(x,y),(x+w,y+w),(255,0,0),10)
        
    return face_img
```

```python
result = detect_face(nadia)
```

```python
plt.imshow(result, cmap='gray')
```

```python
def detect_face(img):
    
    face_img = img.copy()
    
    face_rects = face_cascade.detectMultiScale(face_img,scaleFactor=1.2,minNeighbors=5)
    
    for (x,y,w,h) in face_rects:
        
        cv2.rectangle(face_img,(x,y),(x+w,y+w),(255,0,0),10)
        
    return face_img
```

```python
result = detect_face(solvay)
plt.imshow(result, cmap='gray')
```

```python
eye_cascade = cv2.CascadeClassifier(r'C:/Users/Gaya/Desktop/Computer-Vision-with-Python/DATA/haarcascades/haarcascade_eye.xml')
```

```python
def eye_face(img):
    
    face_img = img.copy()
    
    eyes_rects = eye_cascade.detectMultiScale(face_img)
    
    for (x,y,w,h) in eyes_rects:
        
        cv2.rectangle(face_img,(x,y),(x+w,y+w),(255,0,0),10)
        
    return face_img
```

```python
result = eye_face(nadia)
plt.imshow(result, cmap='gray')
```

```python
cap = cv2.VideoCapture(0)

while True:
    
    ret,frame = cap.read(0)
    
    frame = detect_face(frame)
    
    cv2.imshow('Video Face Detect', frame)
    
    k = cv2.waitKey(1)
    
    if k==27:
        break

cap.release()
cv2.destroyAllWindows()
```
