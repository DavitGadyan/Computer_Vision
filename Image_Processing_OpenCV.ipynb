# Blending images

```python
img1 = cv2.imread('00-puppy.jpg')
img2 = cv2.imread('watermark_no_copy.png')
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
print(img1.shape)
print(img2.shape)
```

```python
img1 = cv2.resize(img1,(1200,1200))
img2 = cv2.resize(img2,(1200,1200))
```

```python
print(img1.shape)
print(img2.shape)
```

```python
blended = cv2.addWeighted(src1=img1,alpha=1,src2=img2,beta=0.2,gamma=20)
```

```python
plt.imshow(blended)
plt.show()
```

# Reassignment

```python
img1 = cv2.imread('00-puppy.jpg')
img2 = cv2.imread('watermark_no_copy.png')
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
img2 = cv2.resize(img2,(300,300))
```

```python
print(img1.shape)
print(img2.shape)
```

```python
larger_img = img1
smaller_img = img2
```

```python
y_offset = 0
x_offset = 0
```

```python
y_end = y_offset + smaller_img.shape[1]
x_end = x_offset + smaller_img.shape[0]
```

```python
larger_img[y_offset:y_end,x_offset:x_end] = smaller_img
```

```python
plt.imshow(larger_img)
plt.show()
```
